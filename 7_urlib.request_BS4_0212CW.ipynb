## Множества set()

Это нечто среднее между списком и словарем (одномерный словарь). Как в словарях, 
во множествах нет понятия о последовательности и нет повторов элементов.

Множества – это как словарь без значений. Не работает то, что работает со словарями. Со множествами не работает append()

Можно превратить список во множество и наоборот.

```python
l = [1,1,1,2,3,3]
s1 = set(l)
print(s1)

ls = list(s1)
print(ls)
```

Можно находить объединения и пересечения во множествах, вычитать их,
убрать пересечения.


```python
s2 = {3,4,5}
s3 = s1 | s2 #черта – логическое ИЛИ, здесь объединение

print(s3)

s4 = s1 & s2 #пересечение
print(s4)

s5 = s2 - s1 #остались только те элементы, которых нет в s1
print(s5)

s6 = s1 ^ s2 #убрать пересечение
print(s6)
```

URL https://lenta.ru

Если жулик зашифрует свой сайт, то этот сайт не получится расшифровать так же, как сайт

Информация про код ответа сайта (если 200, значит, все хорошо)

```python
import urllib.request

r = urllib.request.urlopen("http://lenta.ru") #сюда получаем ответ, 200, код ответа
content = r.read() #читаем файл, вот данные с сайта, тип - байты
print(type(content)) #<class 'bytes'>
```

Контент таким образом будет в байтах, а нам надо раскодировать информацию, чтобы получился текст

```python
html = content.decode("utf-8")
print(html)
```

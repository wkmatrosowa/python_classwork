Json – хитрый способ для хранения данных

Есть простые способы хранения данных – например, таблицы csv, tsv

Не все данные могут быть в виде таблице. Бывают сложные структуры, которые в таблицы не помещаются

Суть json: возможность записывать сложную структуру, не подходящую под формат таблицы, например, для словаря с большим количеством вложенности

{“k”:{“k1”:{“k2”:””””}}}

Специальный модуль для работы с json в питоне (не поддерживает множества)

```python
import json

d = {"k": [1, 2, 3]}
with open("1.json", "w") as fj:
    json.dump(d, fj, ensure_ascii=False, indent=2) #fj – дескриптор файла
```

Детали:

1. По умолчанию json записывает буквы из кириллицы в нечеловеческом виде (не "я", а набор символов).
Это не всегда удобно. Чтобы это исправить, задаем дополнительный аргумент функции ensure_ascii.

2. json файл по умолчанию записывается в одну строчку, что мешает понятности человеком. Есть возможность сделать так, чтобы json файл записывался немного по-другому (с отступами):
использовать indent=число отступов
{“k”:
[1,
2,
3]
}

На случай, если мы хотим с диска сгрузить обратно:

```python
with open("1.json") as f:
    json.load(f)
    print(d)
```

Без ключа доступа (access token) ничего может не сработать. Этот токен выдается нам как пользователям специально для общения с сервером

Сервер, просканировав наш запрос, может не отдать информацию, если мы не предоставим access token.

Вся хитрость в том, чтобы получить где-то этот access token.
